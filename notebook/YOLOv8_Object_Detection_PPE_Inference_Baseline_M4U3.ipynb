```python
!pip install ultralytics -q
import os
from ultralytics import YOLO

# Define the path to your uploaded weights
model_path = '/content/best.pt'

if os.path.exists(model_path):
    # Load the trained model
    model = YOLO(model_path)
    print("✅ Model loaded successfully from session storage!")
else:
    print("❌ Error: 'best.pt' not found. Please upload it using the folder icon on the left.")
```

```python
# 1. Specify the image you uploaded to the sidebar
# Change 'test_image.jpg' to match your filename!
image_path = '/content/test_image.jpg'

if os.path.exists(image_path):
    # 2. Run the "augmented eyes" (inference)
    # We use a confidence threshold of 0.25 to catch most PPE
    results = model.predict(source=image_path, conf=0.25)

    # 3. Visualize the detections
    for r in results:
        # Plot the boxes and labels onto the image
        res_plotted = r.plot()

        # Show the final result in the notebook
        print(f"🔍 Detections for {image_path}:")
        cv2_imshow(res_plotted)
else:
    print(f"❌ Error: '{image_path}' not found in the sidebar. Please upload it!")
```
